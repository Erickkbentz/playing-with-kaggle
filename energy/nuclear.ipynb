```python
import pandas as pd
import matplotlib.pyplot as plt
import os

DATA_DIR = "data"

RATES_DEATHS_CSV = "rates_death_from_energy_production_per_twh.csv"
ENTITY = "Entity"
DEATHS = "Deaths per TWh of electricity production"
```

```python
df = pd.read_csv(os.path.join(DATA_DIR, RATES_DEATHS_CSV))
entities = df[ENTITY]
deaths = df[DEATHS]

df
```

```python
ax = df.plot.bar(x=ENTITY, y=DEATHS, legend=True)
```

```python
GLOBAL_DATA = "global_power_plant_database.csv"

df = pd.read_csv(os.path.join(DATA_DIR, GLOBAL_DATA))
df
```

```python
df[[
    "primary_fuel", 
    "estimated_generation_gwh_2013", 
    "estimated_generation_gwh_2014", 
    "estimated_generation_gwh_2015", 
    "estimated_generation_gwh_2016",
    "estimated_generation_gwh_2017"
]]
```

```python
# Group by fuel type and sum the growths
growth_by_fuel = df.groupby('primary_fuel')[[
    'estimated_generation_gwh_2013', 
    'estimated_generation_gwh_2014', 
    'estimated_generation_gwh_2015', 
    'estimated_generation_gwh_2016',
    'estimated_generation_gwh_2017'
    ]].sum()

growth_by_fuel
```

```python
# Plot the results
growth_by_fuel.plot(kind='bar')

plt.show()
```
